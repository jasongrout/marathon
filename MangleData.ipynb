```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/WaltDisneyWorld-2000.csv', encoding='utf8', sep='\t')
df.columns
```

```python
def get_time(df, newdf, column_name):
    var = column_name.replace(' ','')
    if column_name in df.columns:
        tmp = df[column_name].str.extract('(?P<Hours>\d+):(?P<Minutes>\d+):(?P<Seconds>\d+)').convert_objects(convert_numeric=True)
        newdf[var + 'TotalMinutes'] = tmp['Hours']*60+tmp['Minutes']
        newdf[var + 'Seconds'] = tmp['Seconds']
    else:
        newdf[var + 'TotalMinutes'] = np.nan
        newdf[var + 'Seconds'] = np.nan

def get_div(df, newdf):
    if 'DIV' in df.columns:
        newdf[['DivGender', 'DivBottom', 'DivTop']] = df['DIV'].str.extract(
            '(?P<DivGender>\w)(?P<DivBottom>\d+)-(?P<DivTop>\d+)')
    else:
        newdf['DivGender'] = np.nan
        newdf['DivBottom'] = np.nan
        newdf['DivTop'] = np.nan
def get_bq(df, newdf):
    if 'BQ*' in df.columns:
        newdf['BQ'] = (df['BQ*']=="BQ").astype(int)
    else:
        newdf['BQ'] = np.nan

def get_city(df, newdf):
    if ' City, State, Country' in df.columns:
        newdf['Origin'] = df[' City, State, Country']
    else:
        newdf['Origin'] = np.nan

newdf = df[['race','year','OverAllPlace']]
newdf[['Name', 'Gender', 'Age']] = df['Last Name, First Name(Sex/Age)'].str.extract('(?P<Name>[^(]+)\((?P<Gender>[MF])(?P<Age>\d+)?\)')
newdf['Name'] = newdf['Name'].map(unicode.strip)
get_time(df, newdf, 'Time')
get_time(df, newdf, 'Net Time')
get_div(df, newdf)
get_bq(df, newdf)
```
